# SVD using tensorFlow

```python
import numpy as np
import pandas as pnd
import tensorflow as tf
```

```python
def cosine_similarity(a, b):
    '''
    This function will calculate the cosine similarity between two vectors
    '''
    return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))
```

```python
dataset = pnd.read_csv("user_song_ratings.csv", index_col=0)
dataset
```

```python
pre_processed_dataset = dataset.replace({0: 1})
```

```python
processed_dataset = pre_processed_dataset.drop(["listen_count"], axis=1)
processed_dataset.sort_values('song_id', inplace=True)
processed_dataset
```

```python
pivot_table = pnd.pivot_table(processed_dataset, values='ratings', index='song_id', columns='user_id')
```

```python
back_up = pivot_table
```

```python
back_up = back_up.where(pivot_table.notna(), np.round(pivot_table.mean(axis=1)), axis=0)
```

```python
back_up
```

```python
np_matrix = back_up.sort_values('song_id').to_numpy()
```

```python
print(np_matrix.T.shape)
```

### Take np matrix to tensorFlow tensor

```python
tf_tensor = tf.convert_to_tensor(np_matrix.T[:1000, :5000], dtype=tf.float32)
```

```python
tf_tensor
```

### Create the SVD

```python
s, u, v = tf.linalg.svd(tf_tensor, full_matrices=True)
```

```python
u
```

### Truncate SVD to our components

```python
num_components = 50
```

#### trucate U and V and Sigma

```python
v.numpy().shape
```

```python
u_truncate = u[:, :num_components]
v_truncate = v[:num_components, :]
s_truncate = tf.linalg.diag(s)[:num_components, :num_components]
# u_truncate = u[:num_components, :]
# v_truncate = v[:num_components, :]
# s_truncate = tf.linalg.diag(s)[:num_components, :num_components]
```

#### Calculate distance matrix with scipy
